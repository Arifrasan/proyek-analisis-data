### Cleaning Data

Mengubah tipe data dteday dari object menjadi datetime

```python
#Mengubah tipe data dteday
day_df["dteday"] = pd.to_datetime(day_df["dteday"])
hour_df["dteday"] = pd.to_datetime(hour_df["dteday"])
```

Mengubah nama kolom untuk memperjelas makna

```python
# Mengubah nama kolom tabel day
day_df.rename(columns={
    'dteday': 'dateday',
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weather_cond',
    'cnt': 'count'
}, inplace=True)

# Mengubah nama kolom tabel hour
hour_df.rename(columns={
    'dteday': 'dateday',
    'yr': 'year',
    'hr': 'hour',
    'mnth': 'month',
    'weathersit': 'weather_cond',
    'cnt': 'count'
}, inplace=True)
```

Mengubah keterangan season dari angka menjadi nama musim

```python
#mengubah 1,2,3,4 menjadi 'spring', 'summer', 'fall', 'winter'
day_df.season.replace((1,2,3,4), ('Spring','Summer','Fall','Winter'), inplace=True)
hour_df.season.replace((1,2,3,4), ('Spring','Summer','Fall','Winter'), inplace=True)
```

Mengubah keterangan angka di weekday menjadi nama hari

```python
# mengubah nilai weekday menjadi hari berdasarkan tanggal pada dteday
day_df['weekday'] = day_df['dateday'].dt.day_name()
hour_df['weekday'] = hour_df['dateday'].dt.day_name()
```

Mengubah keterangan angka pada year menjadi tahun sebenarnya

```python
day_df['year'] = day_df['dateday'].dt.year
hour_df['year'] = hour_df['dateday'].dt.year
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana perbanding penyewaan sepeda setiap harinya?


```python
# Membuat subplot dengan 1 baris dan 1 kolom, dengan ukuran (15, 6)
fig, ax = plt.subplots(figsize=(15, 6))

# Membuat barplot untuk y="count" dan x="weekday", menggunakan data=day_df
sns.barplot(
    y="count",
    x="weekday",
    data=day_df,
    hue="weekday",
    palette={"Friday": "#FF3D9A", "Monday": "#F8DB46", "Saturday": "#F8DB46", "Sunday": "#F8DB46", "Thursday": "#F8DB46", "Tuesday": "#F8DB46", "Wednesday": "#F8DB46"},
    ax=ax,
    order=["Friday", "Monday", "Saturday", "Sunday", "Thursday", "Tuesday", "Wednesday"],
    legend=False
)

# Mengatur judul, label y dan x
ax.set_title("Grafik Antar Hari", loc="center", fontsize=20)
ax.set_ylabel("Jumlah Penyewaan Sepeda", fontsize=15)
ax.set_xlabel("Hari", fontsize=15)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

# Menampilkan plot
plt.show()
```

### Pertanyaan 2: Bagaimana tren pesepada casual dan registered pada tahun 2011 dan 2012?

```python
# Hitung jumlah pesepeda registered dan casual setiap tahunnya
result_df = day_df.groupby(by="year").agg({
    "registered": ["sum"],
    "casual": ["sum"]
}).reset_index()

# Visualisasi menggunakan clustered bar chart
plt.figure(figsize=(10, 6))

bar_width = 0.35
bar_positions_1 = range(len(result_df))
bar_positions_2 = [pos + bar_width for pos in bar_positions_1]

plt.bar(bar_positions_1, result_df[("registered", "sum")], width=bar_width, label='Registered')
plt.bar(bar_positions_2, result_df[("casual", "sum")], width=bar_width, label='Casual')

plt.xlabel('Year')
plt.ylabel('Total Count')
plt.title('Total Registered and Casual Bike Rentals Each Year')
plt.xticks([pos + bar_width/2 for pos in bar_positions_1], result_df['year'])
plt.legend()

plt.show()
```

## Conclusion

- Conclution pertanyaan 1 : Dari bar chart di atas dapat diketahui bahwa penyewaan sepeda paling banyak dilakukan pada hari jumat(friday) dan yang paling sedikit pada hari minggu(sunday) dengan jarak yang tidak terlalu besar antar harinya.
- Conclution pertanyaan 2 : Dari clustered bar chart di atas dapat diketahui bahwa tren pesepeda registered naik lebih dari setengah di tahun 2012, sedangkan pesepeda casual juga mengalami tren kenaikan lebih dari setengah tahun 2012.
